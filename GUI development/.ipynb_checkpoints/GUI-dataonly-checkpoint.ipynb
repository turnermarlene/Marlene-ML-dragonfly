```python
import tkinter as tk
from dragonfly import maximise_function, minimise_function
import socket
import struct
import numpy as np
import matplotlib.pyplot as plt
import time
import re
from functions_hex_cam import *

N = 3

server_address_probecampost = ('192.168.15.24', 65280)
server_address_probecampostnear = ('192.168.15.25', 65264)
hexapod_address = ('192.168.15.16', 65529)

subcriptionstring = b'Wait>>MeanCounts'
SubcriptionCmdLength = len(subcriptionstring)
sizepack = struct.pack('>i', SubcriptionCmdLength)

subcriptionstring2 = b'Wait>>Centroidx,Centroidy'
SubcriptionCmdLength2 = len(subcriptionstring2)
sizepack2 = struct.pack('>i', SubcriptionCmdLength2)

udp_client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
```

```python
def initialise():
    var.set('Messages: Getting centroids')
    centroidupdatemessage.set('In Progress')
    master.update()
    
    
    hexy_temp = get_current_hex_pos('ypos')
    hexz_temp = get_current_hex_pos('zpos')
    hexw_temp = get_current_hex_pos('wangle')
    hexv_temp = get_current_hex_pos('vangle')
    
    update_hex_pos(20.0,float(e_hexz.get()),float(e_wangle.get()),float(e_vangle.get()))
    
    cenx1,ceny1 = get_cam_centroid(server_address_probecampostnear)
    cenx2,ceny2 = get_cam_centroid(server_address_probecampost)         
    
    e_cen1x.insert(0,np.round(cenx1,1))
    e_cen1y.insert(0,np.round(ceny1,1))
    e_cen2x.insert(0,np.round(cenx2,1))
    e_cen2y.insert(0,np.round(ceny2,1))
    
    update_hex_pos(float(e_hexy.get()),float(e_hexz.get()),float(e_wangle.get()),float(e_vangle.get()))
    
    var.set('Messages: Done with centroids')
    centroidupdatemessage.set(' ')
    Initialise.configure(bg='green')
    master.update()
    return

```

```python
def update_hex_pos(hexy,hexz,wangle,vangle):
    udp_client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    hexy = np.round(hexy,4)
    hexz = np.round(hexz,4)
    wangle = np.round(wangle,4)
    vangle = np.round(vangle,4)
    
    hexycommand =bytes("setypos>>"+str(hexy),'ascii')
    #print(hexycommand)
    hexzcommand =bytes("setzpos>>"+str(hexz),'ascii')
    wanglecommand =bytes("setwangle>>"+str(wangle),'ascii')
    vanglecommand =bytes("setvangle>>"+str(vangle),'ascii')
    
    udp_client.sendto(hexycommand, hexapod_address)
    ypos_m = get_current_hex_pos('ypos')
    while np.abs(ypos_m-hexy) > 0.01:
        ypos_m = get_current_hex_pos('ypos')
    time.sleep(1)

    udp_client.sendto(hexzcommand, hexapod_address)
    zpos_m = get_current_hex_pos('zpos')
    while np.abs(zpos_m-hexz) > 0.1:
        zpos_m = get_current_hex_pos('zpos')
    time.sleep(1)
    
    udp_client.sendto(wanglecommand, hexapod_address)
    wpos_m = get_current_hex_pos('wangle')
    while np.abs(wpos_m-wangle) > 0.01:
        wpos_m = get_current_hex_pos('wangle')
    time.sleep(1) 
    
    udp_client.sendto(vanglecommand, hexapod_address)
    vpos_m = get_current_hex_pos('vangle')
    while np.abs(vpos_m-vangle) > 0.01:
        vpos_m = get_current_hex_pos('vangle')
    #time.sleep(1)   
    
    return


def get_cam_mean(cam_server_adress): #select camera
    client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client.connect(cam_server_adress)
    client.sendall(sizepack + subcriptionstring)
    
    #initialise loop
    counts = np.zeros(N)
    i = 0
    
    while i < N:
        size = struct.unpack('>i', client.recv(4))[0]  # Extract the msg size from four bytes - mind the encoding
        str_data = client.recv(size)

        data = re.findall(r'\d+', str_data.decode('ascii'))
        counts[i] = np.int(data[2])
        #print('Data size: %s Data value: %s' % (size, str_data.decode('ascii')))
        i = i+1
    client.close()
    
    meancounts_campost = np.mean(counts) 
    
    return meancounts_campost

def get_cam_centroid(cam_server_adress): #select camera
    client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client.connect(cam_server_adress)
    client.sendall(sizepack2 + subcriptionstring2)
    
    #initialise loop
    centroid = np.zeros([N,2])
    i = 0
    
    while i < N:
        size = struct.unpack('>i', client.recv(4))[0]  # Extract the msg size from four bytes - mind the encoding
        str_data = client.recv(size)

        data = re.findall(r'\d+', str_data.decode('ascii'))
        #print(data)
        centroid[i,0] = np.int(data[2])
        centroid[i,1] = np.int(data[4])
        #print('Data size: %s Data value: %s' % (size, str_data.decode('ascii')))
        i = i+1
    client.close()
    
    cenx = np.mean(centroid[:,0]) 
    ceny = np.mean(centroid[:,1]) 
    
    return [cenx,ceny]

def get_current_hex_pos(parameter):
    client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client.connect(hexapod_address)

    subcriptionstring = bytes("Wait>>"+parameter,'ascii')
    #print(subcriptionstring)
    SubcriptionCmdLength = len(subcriptionstring)
    sizepack = struct.pack('>i', SubcriptionCmdLength)


    client.sendall(sizepack + subcriptionstring)
    size = struct.unpack('>i', client.recv(4))[0]  # Extract the msg size from four bytes - mind the encoding
    str_data = client.recv(size)
    #print('Data size: %s Data value: %s' % (size, str_data.decode('ascii')))
    position = np.float(re.findall(r'[-+]?\d+\.\d+', str_data.decode('ascii'))[0])
    client.close()
    
    return position

def opt_align(hexp):
    hexy = hexp[0]
    hexz = hexp[1]
    wangle = hexp[2]
    vangle = hexp[3]

    #update hexapod pos
    update_hex_pos(hexy,hexz,wangle,vangle)
    
    #get cam property
    meancounts_campost1 = get_cam_mean(server_address_probecampostnear)
    meancounts_campost2 = get_cam_mean(server_address_probecampost)
    
    meancounts_campost = meancounts_campost1+meancounts_campost2
    
    if meancounts_campost>25:
        cenx1,ceny1 = get_cam_centroid(server_address_probecampostnear)
        cenx2,ceny2 = get_cam_centroid(server_address_probecampost)
        
        #get centroids when cap is out from gui
        capout_cenx1 = float(e_cen1x.get())
        capout_ceny1 = float(e_cen1y.get())
        capout_cenx2 = float(e_cen2x.get())
        capout_ceny2 = float(e_cen2y.get())
    
        cen = 6*np.sqrt((np.abs(cenx1-capout_cenx1)+np.abs(ceny1-capout_ceny1))**2)+np.sqrt((np.abs(cenx2-capout_cenx2)+np.abs(ceny2-capout_ceny2))**2)
    else:
        cen=2000-meancounts_campost*2
    
    print(np.round(cen))
    return cen

def optimise_coarse():
    #get measurement range from gui
    var.set('Messages: optimising')
    prealignmentupdatemessage.set('In progress')
    master.update()
    
    r_hexy = float(e_hexy.get())
    r_hexy_r = float(e_hexy_r.get())
    r_hexz = float(e_hexz.get())
    r_hexz_r = float(e_hexz_r.get())
    r_hexw = float(e_wangle.get())
    r_hexw_r = float(e_wangle_r.get())
    r_hexv = float(e_vangle.get())
    r_hexv_r = float(e_vangle_r.get())
                     
    domain = [[r_hexy-r_hexy_r,r_hexy+r_hexy_r],[r_hexz-r_hexz_r,r_hexz+r_hexz_r],[r_hexw-r_hexw_r,r_hexw+r_hexw_r],[r_hexv-r_hexv_r,r_hexv+r_hexv_r]]
                     
    max_capital = Capital.get()
    
    max_val, max_pt, history = minimise_function(opt_align, domain, max_capital,opt_method='bo')

    #finsish set to optimum values
    print('We are done here:')
    print(max_val)
    print(max_pt)
    opt_align(max_pt)
    
    prealignmentupdatemessage.set(' ')
    var.set('Messages: optimisation is done')
    PreAlign.configure(bg='green')
    master.update()
    
def test():
    print(Capital.get())
```

```python
#define GUI shape, layout and input

master = tk.Tk()
master.title('ML Capillary Alignment')
master.geometry("700x300")
master.option_add( "*font", "Calibri 10" )

prealignmentupdatemessage= tk.StringVar()
message_display_pre = tk.Label(master, textvariable=prealignmentupdatemessage, width=70, anchor='w',fg='red')
prealignmentupdatemessage.set(' ')
message_display_pre.place(x=100,y=230)
PreAlign = tk.Button(master, text="Pre-Alignment",command=optimise_coarse,bg='red')
PreAlign.place(x=100,y=250)

PreAlign = tk.Button(master, text="Alignment",bg='red')
PreAlign.place(x=200,y=250)

#Camera centroid input

tk.Label(text='Starting Point', width=15).grid(row=0,column=1)
tk.Label(text='Range', width=15).grid(row=0,column=2)
tk.Label(text='Hex y: ', width=15).grid(row=1,column=0)
tk.Label(text='Hex z: ', width=15).grid(row=2,column=0)
tk.Label(text='Wangle: ', width=15).grid(row=3,column=0)
tk.Label(text='Vangle: ', width=15).grid(row=4,column=0)

hexy_prefill = get_current_hex_pos('ypos')
hexz_prefill = get_current_hex_pos('zpos')
hexw_prefill = get_current_hex_pos('wangle')
hexv_prefill = get_current_hex_pos('vangle')

e_hexy = tk.Entry(master)
e_hexy.insert(0,hexy_prefill)
e_hexy.grid(row=1, column=1)

e_hexz = tk.Entry(master)
e_hexz.insert(0,hexz_prefill)
e_hexz.grid(row=2, column=1)

e_wangle = tk.Entry(master)
e_wangle.insert(0,hexw_prefill)
e_wangle.grid(row=3, column=1)

e_vangle = tk.Entry(master)
e_vangle.insert(0,hexv_prefill)
e_vangle.grid(row=4, column=1)

e_hexy_r = tk.Entry(master)
e_hexy_r.insert(0,0.2)
e_hexy_r.grid(row=1, column=2)

e_hexz_r = tk.Entry(master)
e_hexz_r.insert(0,0.2)
e_hexz_r.grid(row=2, column=2)

e_wangle_r = tk.Entry(master)
e_wangle_r.insert(0,0.15)
e_wangle_r.grid(row=3, column=2)

e_vangle_r = tk.Entry(master)
e_vangle_r.insert(0,0.15)
e_vangle_r.grid(row=4, column=2)

tk.Label(text='N. Cam: ', width=15).grid(row=5,column=0)
Ncamimg = tk.Entry(master)
Ncamimg.insert(0,3)
Ncamimg.grid(row=5, column=1)

tk.Label(text='Capital: ', width=15).grid(row=6,column=0)
Capital = tk.Entry(master)
Capital.insert(0,25)
Capital.grid(row=6, column=1)

tk.Label(text='Centroid x', width=15).grid(row=7,column=1)
tk.Label(text='Centroid y', width=15).grid(row=7,column=2)

tk.Label(text='Cam 1: ', width=15).grid(row=8,column=0)
e_cen1x = tk.Entry(master)
e_cen1x.grid(row=8, column=1)

e_cen1y = tk.Entry(master)
e_cen1y.grid(row=8, column=2)

tk.Label(text='Cam 2: ', width=15).grid(row=9,column=0)
e_cen2x = tk.Entry(master)
e_cen2x.grid(row=9, column=1)

e_cen2y = tk.Entry(master)
e_cen2y.grid(row=9, column=2)

centroidupdatemessage= tk.StringVar()
message_display_cen = tk.Label(master, textvariable=centroidupdatemessage, width=70, anchor='w',fg='red')
centroidupdatemessage.set(' ')
message_display_cen.place(x=420,y=155)
Initialise = tk.Button(master, text="Get Centroids",command=initialise, bg='red')
Initialise.place(x=420,y=175)


#Close GUI Button
button_close = tk.Button(master, text="Close", command=master.destroy,bg='pink')
button_close.place(x=420,y=250)


var= tk.StringVar()
message_display = tk.Label(master, textvariable=var, width=110, relief=tk.RIDGE,anchor='w')
var.set('Messages:')
message_display.place(x=0,y=280)
```
